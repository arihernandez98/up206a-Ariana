```python
columns_to_keep = ['geoid',
 'name',
 'B25003001',
 'B25003002',
 'B25003003', 
 'geometry'] 
```

I got rid of the Errors and kept the relevant data: 
B25003001 = Total
B25003002 = Own
B25003003 = Rent

```python
tenure = tenure [columns_to_keep]
```

Updated the tenure to only include the columns I wanted to keep. 

```python
tenure.head()
```

```python
tenure.columns = ['geoid',
 'name',
 'Total',
 'Own',
 'Rent',
 'geometry']
```

I renamed the housing tenure columns to reflect what they actually are. 

```python
tenure.head()
```

```python
tenure['Percent Own'] = tenure['Own']/tenure['Total']*100
tenure['Percent Rent'] = tenure['Rent']/tenure['Total']*100
```

I normalized my data by creating percentages of ownership and renters in LA city. 

```python
tenure.head()
```

```python
tenure.plot(figsize=(12,12),
                 column='Percent Rent',
                 legend=True, 
                 scheme='NaturalBreaks')
```

This is my first chart of renters in LA City. However, with a basemap and a better colorscheme this would be a nicer, easier to read map. 

```python
tenure.plot(figsize=(12,12),
            column='Percent Rent', 
            cmap='Pastel1', 
            scheme='user_defined',
            classification_kwds={'bins':[25, 50, 75, 100]},
            edgecolor='white', 
            linewidth=0.1, 
            legend=True
           )
```

First, I added some color to the map. This color scheme isn't exactly fitting, but I'll fix it in the next map.

```python

fig, axs = plt.subplots(1, 2, figsize=(15, 12), sharex=True, sharey=True)


ax1, ax2 = axs

tenure.plot(column='Percent Rent', 
            cmap='Accent', 
            scheme='user_defined',
            classification_kwds={'bins':[25, 50, 75,100]},
            edgecolor='white', 
            linewidth=0.1, 
            ax=ax1,
            legend=True
           )

ax1.axis("off")
ax1.set_title("Percent Rent")


tenure.plot(column='Percent Own', 
            cmap='Accent', 
            scheme='user_defined',
            classification_kwds={'bins':[25, 50, 75, 100]},
            edgecolor='white', 
            linewidth=0.1, 
            ax=ax2, # this assigns the map to the subplot
            legend=True
           )

ax2.axis("off")
ax2.set_title("Percent Own")
```

I made two maps, one of percent renters and the other of percent homeowners. I think if I had more variables this map would be helpful but since there are only two options and the sections are split by 25/50/75/100 the maps are an inverse of one another. 

```python
tenure_web_mercator = tenure.to_crs(epsg=3857) 
fig, ax = plt.subplots(figsize=(15, 15))

 
tenure_web_mercator[tenure_web_mercator['Percent Rent'] > 90].plot(                                      
                                                               ax=ax,
                                                               color="red",
                                                               alpha=0.8
)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Renters in LA City',fontsize=24,pad=20)

# add a basemap
ctx.add_basemap(ax)
```

This map shows areas of LA where over 90% of people rent their homes. I chose this basemap because you can see the green in the background and unsurpisingly areas with over 90% renters seem to be in urban areas and far away from green space. The map would look better with a boundary around LA City, but I'm not quite sure how to do that...yet. 

```python
tenure_web_mercator = tenure.to_crs(epsg=3857) 
fig, ax = plt.subplots(figsize=(15, 15))

 
tenure_web_mercator[tenure_web_mercator['Percent Rent'] > 90].plot(                                      
                                                               ax=ax,
                                                               color="blue",
                                                               alpha=0.8
)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Renters in LA City',fontsize=24,pad=20)

# add a basemap
ctx.add_basemap(ax, source=ctx.providers.Stamen.Watercolor)
```

I also wanted to look at a watercolor basemap because it is pretty but also because it does a great job of showing green spaces. I changed the color to represent percent renters to blue because it had a higher contrast to the warmer colors on the map.
